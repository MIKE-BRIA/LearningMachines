## Operators

```python
## Arithmetic operations

a = 10
b = 5
print("Addition:", a + b)
print("Subtraction:", a - b)
print("Multiplication:", a * b)
print("Division:", a / b)
print("Floor Division:", a // b)
print("Modulus:", a % b)
print("Exponentiation:", a ** b)
```

```python
## comparison operators

a = 10
b = 15

print('results of comparison operators:\n')
print('Equal to operator','\na == b:', a == b, )
print('Not Equal to operator','\na != b:', a != b)
print('Greater than operator','\na > b:', a > b)
print('Less than operator','\na < b:', a < b)
print('Greater than or equal to operator','\na >= b:', a >= b)
print('Less than or equal to operator','\na <= b:', a <= b)
```

```python
## Logical operators

## And, or, not

x = True
y = False

result_and = x and y
result_or = x or y
result_not_x = not x
result_not_y = not y

print("x and y:", result_and)
print("x or y:", result_or)
print("not x:", result_not_x)
print("not y:", result_not_y)
```
