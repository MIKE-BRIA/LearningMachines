## Conditional statements

```python
## if statement
## are used to make conditional decisions in code

age = 18

if age >= 18:
    print("You are an adult.")
```

```python
## else statement
## provides an alternative block of code to execute if the condition in the if statement is not met

age = 16

if age >= 18:
    print("You are an adult.")
else:
    print("You are a minor.")
```

```python
## elif statement
## allows you to check multiple conditions sequentially "elif" stands for "else if"


age = 65

if age < 18:
    print("You are a minor.")
elif age < 65:
    print("You are an adult.")
else:
    print("You are a senior citizen.")
```

```python
## Nested conditional statements
## you can have if statements inside other if statements to create more complex decision-making structures

num = int(input("Enter a number: "))

if num > 0:
    print('The number is positive.')
    if num % 2 == 0:
        print('The number is even.')
    else:
        print('The number is odd.')
elif num < 0:
    print('The number is negative.')
else:
    print('The number is zero.')
```

```python
## Practical Example: Grading System
## Determine if a year is a leap year using nested if statements

year = int(input("Enter a year: "))

if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
    print(f"{year} is a leap year.")
else:
    print(f"{year} is not a leap year.")

# think this is the most efficient way to write this code but there are other ways to do it as well.
```

## Loops

```python
## For loops
## A for loop is used to iterate over a sequence (like a list, tuple, string, or range) and execute a block of code for each item in the sequence.
## Range is a built-in function

for i in range(5):
    print(i)
```

```python
for i in range(1, 11):
    print(i)
```

```python
range(1,18,2)
```

```python
for i in range(1,18,2):
    print(i)
```

```python
## Strings with for loops

str = 'Brian Michael'

for i in str:
    print(i)
```

```python
## While Loops
## A while loop repeatedly executes a block of code as long as a specified condition is true

count = 0

while count < 5:
    print(count,'We are still not there yet!')
    count += 1
```

```python
## Loop control statements
## break statement: used to exit a loop prematurely when a certain condition is met

## Break statement example

for i in range(10):
    if i == 5:
        break
    print(i)
```

```python
## Continue statement: used to skip the current iteration of a loop and move to the next iteration
for i in range(10):
    if i % 2 == 0:
        continue
    print(i)
```

```python
## Pass statement: used as a placeholder in loops or conditional statements when you want to indicate that no action is required

for i in range(5):
    if i == 2:
        pass
    print(i)
```

```python
## Nested Loops: a loop inside another loop

for i in range (3):
    for j in range(2):
        print(f"Outer loop iteration {i}, Inner loop iteration {j}")
```

```python
 ## Example : Calculate the sum of first N natural numbers using a while and for loop

 ## While loop example

n=10
sum = 0
count = 1

while count <= n:
    sum += count
    count += 1
    print(n, sum, count)
print(f"The sum of the first {n} natural numbers is: {sum}")
```

```python
sum = 0

for i in range(1, 11):
    sum += i
    print(n, sum, i)
print(f"The sum of the first {n} natural numbers is: {sum}")
```

```python
## Example: Prime numbers between 1 and 100

for num in range(1, 101):
    if num > 1:
        for i in range(2, num):
            if (num % i) == 0:
                break
        else:
            print(num,'is a prime number.')
```
